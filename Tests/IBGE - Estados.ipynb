```python
import ibge
import pandas as pd
import json
```

```python

from ibge.localidades import *
estados = Estados()
```

```python
df = pd.DataFrame(  pd.json_normalize(estados.json()), columns=['sigla', 'nome', 'regiao.nome'])
```

```python
df.groupby('regiao.nome')['regiao.nome'].value_counts()
```

```python
df.groupby('nome')['regiao.nome'].value_counts().unstack().fillna(0)
```

```python
df.groupby('regiao.nome')['regiao.nome'].value_counts().plot(kind='pie')
```
